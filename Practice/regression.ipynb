```python
import numpy as np
import matplotlib.pyplot as plt


x = [1.1, 1.3, 1.7, 2.0, 2.4, 2.9, 3.2, 3.4, 3.5, 3.7, 4.1, 4.3, 4.3, 4.4, 4.7, 5.1,
        5.3, 5.5, 6.2, 6.3, 6.8, 7.3, 7.9, 8.4, 8.7, 9.2, 9.5, 9.8, 10.3, 10.7]

y = [35010.0, 42500.0, 39800.0, 41050.0, 42100.0, 53200.0, 58700.0, 50200.0, 60500.0, 59000.0,
        60100.0, 58000.0, 59500.0, 61000.0, 63000.0, 70500.0, 68200.0, 79000.0, 85000.0, 88000.0,
        94000.0, 96500.0, 105000.0, 109000.0, 112000.0, 110000.0, 120000.0, 115000.0, 125000.0, 128000.0]

X = np.array(x)
Y = np.array(y)

# Prepare Matrix A

# A = np.vstack([X**2, X, np.ones(len(X))]).T # --> Polynomial
A = np.vstack([x, np.ones(len(x))]).T


# Formula (A^T.A)^-1.A^T.Y = C
    # np.linalg.inv(A) --> inversing matrix
    # np.dot(A, B) --> multiply 2 matrices
    # np.linalg.pinv(A) --> pseudo inversing matrix

C = np.dot(np.linalg.pinv(A), Y)
# C = np.dot(np.dot(np.linalg.inv(np.dot(A.T,A)), A.T), Y) # --> Scratch Way


# Formula Y_Prediction = mx + c

y_pred = C[0]*X + C[1]
# y_pred = C[0]*X**2 + C[1]*X + C[2] # --> Polynomial

print(C)
print(y_pred)

plt.plot(x, y, "ob")
plt.plot(x, y_pred, "y-")
plt.title("Linear Regression")
plt.legend(["Data Point", "Bestfit Line"])
plt.show()
```
