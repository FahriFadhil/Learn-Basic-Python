```python

import numpy as np

x = [[[10, 2, 1],
      [1, -8, 2],
      [2, 3, -9]],
    [[-12, 2, -3],
     [3, 9, 1],
     [-2, 1, 10]],
    [[11, -1, 3],
     [2, 7, -2],
     [3, 2, 9]],
    [[-10, 5, 1],
     [-4, 11, -3],
     [-2, 3, -8]],
    [[8, 1, -2],
     [-3, 10, 2],
     [-1, -2, 9]]]

y = [[40, 1150, 18],
    [25, -10, 15],
    [30, -10, 12],
    [80, -700, 140],
    [15, 10, -5]]

# z = [[10, 2, 1],
#      [1, -8, 2],
#      [2, 3, -9],
#      [1, -8, 2]]


#? np.array(array) = makes a variable array can use numpy feature
x = np.array(x)
y = np.array(y)

def checkDiag (matrix):
    #? np.abs(num) = return an absolute of the number
    #? np.diag(matrix) = return the trace of a matrix
    #? np.sum(array) = return the sum of an array
    diagonal = np.abs(np.diag(matrix))
    nonDiagonal = np.sum(np.abs(matrix), axis = 1) - diagonal
    # print(diagonal)
    # print(nonDiagonal)

    #? len(array) = return the length of an array
    for i in range (len(diagonal)):
        if(diagonal[i] < nonDiagonal[i]):
            return False
    return True
    
def gaussSeidel (x, y, e = 0.01, n = 50):
    if not checkDiag(x):
        print("Matrix might bot be diagonally dominant")
        return

    #? copy() = duplicate a matrix/array without altering the original
    #? np.zeros(n, n?) = return an array of zeros with the size of n
    #? np.shape(matrix) = return the dimensions of a matrix
    #? np.fill_diagonal(matrix) = return the dimensions of a matrix
    x = x.copy()
    prev_guess = np.zeros(np.shape(x[0]))
    diagonal = np.diag(x).copy()
    np.fill_diagonal(x, 0)

    #? enumerate(array) = make anything iterable (list, array, etc.)
    #? np.dot(matrix, matrix) = return the dot product of 2 arrays
    #? np.sqrt(num) = return square root of a num
    #? np.sqrt(num) = return square root of a num
    for i in range(n):
        new_guess = np.array(prev_guess)
        for j, row in enumerate(x):
            new_guess[j] = (y[j] - np.dot(row, new_guess)) / diagonal[j]
        print(f"iter #{i} = {new_guess}")

        euclidean_distance = np.sqrt(np.dot(new_guess-prev_guess, new_guess-prev_guess))
        prev_guess = new_guess
        if euclidean_distance < e:
            print("Converged")
            return

for i in range(len(x)):
    print(f"Equation #{i + 1}:")
    gaussSeidel(x[i], y[i])
    # print("\n")

```
